```python
import pandas as pd
import numpy as np

data = pd.read_csv("../data/orders_data.csv")
data["date"] = pd.to_datetime(data["date"])
data.sort_values("date", inplace=True)

# --- Feature Engineering (same as Day 4) ---
data["day_of_week"] = data["date"].dt.dayofweek
data["week_of_year"] = data["date"].dt.isocalendar().week.astype(int)
data["month"] = data["date"].dt.month
data["is_weekend"] = data["day_of_week"].isin([5, 6]).astype(int)

data["lag_1"] = data["orders"].shift(1)
data["lag_7"] = data["orders"].shift(7)
data["lag_14"] = data["orders"].shift(14)

data["rolling_7"] = data["orders"].rolling(7).mean()
data["rolling_14"] = data["orders"].rolling(14).mean()

data.dropna(inplace=True)
data.drop(columns=["date"], inplace=True)
data = pd.get_dummies(data, drop_first=True)

X = data.drop("orders", axis=1)
y = data["orders"]
```

```python
import pandas as pd

data = pd.read_csv("../data/orders_data.csv")
data["date"] = pd.to_datetime(data["date"])
data
```

```python
data["day_of_week"] = data["date"].dt.dayofweek
data["month"] = data["date"].dt.month
data["is_weekend"] = data["day_of_week"].isin([5, 6]).astype(int)

data.drop(columns=["date"], inplace=True)
```

```python
data = pd.get_dummies(data, drop_first=True)
data
```

```python
X = data.drop("orders", axis=1)
y = data["orders"]

print(X.shape, y.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)

print("Model trained successfully")
```

```python
predictions = model.predict(X)
predictions
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    shuffle=False
)

print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)
rf.fit(X_train, y_train)
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=5,
    random_state=42
)

xgb.fit(X_train, y_train)
```

```python
models = {
    "Linear Regression": lr,
    "Random Forest": rf,
    "XGBoost": xgb
}
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
results = []

for name, model in models.items():
    y_pred = model.predict(X_test)

    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)

    results.append([name, mae, rmse, r2])
```

```python
import pandas as pd

results_df = pd.DataFrame(
    results,
    columns=["Model", "MAE", "RMSE", "R2 Score"]
)

results_df
```

```python
best_model = results_df.sort_values("RMSE").iloc[0]
best_model
```

```python
import matplotlib.pyplot as plt

best_model_name = best_model["Model"]
best_model_obj = models[best_model_name]

y_pred_best = best_model_obj.predict(X_test)

plt.figure(figsize=(8,5))
plt.plot(y_test.values, label="Actual Orders")
plt.plot(y_pred_best, label="Predicted Orders")
plt.legend()
plt.title(f"Actual vs Predicted Orders ({best_model_name})")
plt.xlabel("Time")
plt.ylabel("Number of Orders")
plt.show()
```

```python
import joblib

joblib.dump(best_model_obj, "final_orders_prediction_model.pkl")
```
